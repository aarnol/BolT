```python
# Define the paths to save the metrics and results
import os
name = "hcpWM_fNIRS_HbC_hcpWM_newMNI_finetuneHBC"
step_metrics_path = os.path.join("..", "Logs", name, "step_metrics.pkl")
epoch_metrics_path = os.path.join("..", "Logs", name, "epoch_metrics.pkl")
test_metrics_path = os.path.join("..", "Logs", name, "test_metrics.pkl")
train_metrics_path = os.path.join("..", "Logs", name, "train_metrics.pkl")
```

```python
import pickle

# Define the path to the pickle file
pickle_path = step_metrics_path

# Load the pickle file
with open(pickle_path, 'rb') as f:
    epoch_metrics = pickle.load(f)
epoch_metrics
```

```python
#plots for accuracy, precision, recall, roc, and loss over time
# Extract metrics from 
import matplotlib.pyplot as plt
import numpy as np



accuracy = [metric['accuracy'] for metric in epoch_metrics]
precision = [metric['precision'] for metric in epoch_metrics]
recall = [metric['recall'] for metric in epoch_metrics]
roc = [metric['roc'] for metric in epoch_metrics]
loss = [metric['loss'] for metric in epoch_metrics]


# Plot accuracy over time
plt.figure(figsize=(12, 6))
plt.subplot(2, 3, 1)
plt.plot(accuracy, label='Accuracy')
plt.xlabel('Step')
plt.ylabel('Accuracy')
plt.title('Accuracy over Time')
plt.legend()

# Plot precision over time
plt.subplot(2, 3, 2)
plt.plot(precision, label='Precision')
plt.xlabel('Step')
plt.ylabel('Precision')
plt.title('Precision over Time')
plt.legend()

# Plot recall over time
plt.subplot(2, 3, 3)
plt.plot(recall, label='Recall')
plt.xlabel('Step')
plt.ylabel('Recall')
plt.title('Recall over Time')
plt.legend()

# Plot ROC over time
plt.subplot(2, 3, 4)
plt.plot(roc, label='ROC')
plt.xlabel('Step')
plt.ylabel('ROC')
plt.title('ROC over Time')
plt.legend()

# Plot loss over time
plt.subplot(2, 3, 5)
plt.plot(loss, label='Loss')
plt.xlabel('Step')
plt.ylabel('Loss')
plt.title('Loss over Time')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# Trim the arrays to the largest multiple of 10
trimmed_size = (len(accuracy) // 10) * 10

avg_accuracy = np.mean(np.array(accuracy[:trimmed_size]).reshape(-1, 10), axis=1)
avg_precision = np.mean(np.array(precision[:trimmed_size]).reshape(-1, 10), axis=1)
avg_recall = np.mean(np.array(recall[:trimmed_size]).reshape(-1, 10), axis=1)
avg_roc = np.mean(np.array(roc[:trimmed_size]).reshape(-1, 10), axis=1)
avg_loss = np.mean(np.array(loss[:trimmed_size]).reshape(-1, 10), axis=1)
```

```python
# Plot averaged metrics over time
%matplotlib inline
plt.figure(figsize=(12, 6))

# Plot averaged accuracy over time
plt.subplot(2, 3, 1)
plt.plot(avg_accuracy, label='Avg Accuracy')
plt.xlabel('Step')
plt.ylabel('Avg Accuracy')
plt.title('Average Accuracy over Time')
plt.legend()

# Plot averaged precision over time
plt.subplot(2, 3, 2)
plt.plot(avg_precision, label='Avg Precision')
plt.xlabel('Step')
plt.ylabel('Avg Precision')
plt.title('Average Precision over Time')
plt.legend()

# Plot averaged recall over time
plt.subplot(2, 3, 3)
plt.plot(avg_recall, label='Avg Recall')
plt.xlabel('Step')
plt.ylabel('Avg Recall')
plt.title('Average Recall over Time')
plt.legend()

# Plot averaged ROC over time
plt.subplot(2, 3, 4)
plt.plot(avg_roc, label='Avg ROC')
plt.xlabel('Step')
plt.ylabel('Avg ROC')
plt.title('Average ROC over Time')
plt.legend()

# Plot averaged loss over time
plt.subplot(2, 3, 5)
plt.plot(avg_loss, label='Avg Loss')
plt.xlabel('Step')
plt.ylabel('Avg Loss')
plt.title('Average Loss over Time')
plt.legend()
plt.suptitle('Averaged Metrics over Time (10 steps)', fontsize=16)
plt.tight_layout()
plt.show()
```

```python
def plot_step_metrics(pickle_path):
    import pickle
    import matplotlib.pyplot as plt
    import numpy as np

    # Load the pickle file
    with open(pickle_path, 'rb') as f:
        epoch_metrics = pickle.load(f)

    accuracy = [metric['accuracy'] for metric in epoch_metrics]
    precision = [metric['precision'] for metric in epoch_metrics]
    recall = [metric['recall'] for metric in epoch_metrics]
    roc = [metric['roc'] for metric in epoch_metrics]
    loss = [metric['loss'] for metric in epoch_metrics]

    # Plot accuracy over time
    plt.figure(figsize=(12, 6))
    plt.subplot(2, 3, 1)
    plt.plot(accuracy, label='Accuracy')
    plt.xlabel('Step')
    plt.ylabel('Accuracy')
    plt.title('Accuracy over Time')
    plt.legend()

    # Plot precision over time
    plt.subplot(2, 3, 2)
    plt.plot(precision, label='Precision')
    plt.xlabel('Step')
    plt.ylabel('Precision')
    plt.title('Precision over Time')
    plt.legend()

    # Plot recall over time
    plt.subplot(2, 3, 3)
    plt.plot(recall, label='Recall')
    plt.xlabel('Step')
    plt.ylabel('Recall')
    plt.title('Recall over Time')
    plt.legend()

    # Plot ROC over time
    plt.subplot(2, 3, 4)
    plt.plot(roc, label='ROC')
    plt.xlabel('Step')
    plt.ylabel('ROC')
    plt.title('ROC over Time')
    plt.legend()

    # Plot loss over time
    plt.subplot(2, 3, 5)
    plt.plot(loss, label='Loss')
    plt.xlabel('Step')
    plt.ylabel('Loss')
    plt.title('Loss over Time')
    plt.legend()

    plt.tight_layout()
    plt.show()

step_metrics_path = r"C:\Users\Alex\Documents\BolT\Analysis\Logs\dynamic15Schaefer_bolT_hcpWM_20241202\step_metrics.pkl"
plot_step_metrics(step_metrics_path)
```

```python
import pickle
import numpy as np

def plot_test_metrics(test_results_path, test_metrics_path):
    import matplotlib.pyplot as plt

    # Load the test results pickle file
    with open(test_results_path, 'rb') as f:
        test_results = pickle.load(f)

    # Load the test metrics pickle file
    with open(test_metrics_path, 'rb') as f:
        test_metrics = pickle.load(f)

    # Extract metrics from test_results and test_metrics
    test_loss = [result['loss'] for result in test_results]
    test_accuracy = [metric['accuracy'] for metric in test_metrics]
    test_precision = [metric['precision'] for metric in test_metrics]
    test_recall = [metric['recall'] for metric in test_metrics]
    test_roc = [metric['roc'] for metric in test_metrics]

    # Plot test metrics
    plt.figure(figsize=(15, 10))

    # Plot test accuracy over time
    plt.subplot(2, 3, 1)
    plt.plot(test_accuracy, label='Test Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.title('Test Accuracy over Time')
    plt.legend()

    # Plot test loss over time
    plt.subplot(2, 3, 2)
    plt.plot(test_loss, label='Test Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.title('Test Loss over Time')
    plt.legend()

    # Plot test precision over time
    plt.subplot(2, 3, 3)
    plt.plot(test_precision, label='Test Precision')
    plt.xlabel('Epoch')
    plt.ylabel('Precision')
    plt.title('Test Precision over Time')
    plt.legend()

    # Plot test recall over time
    plt.subplot(2, 3, 4)
    plt.plot(test_recall, label='Test Recall')
    plt.xlabel('Epoch')
    plt.ylabel('Recall')
    plt.title('Test Recall over Time')
    plt.legend()

    # Plot test ROC over time
    plt.subplot(2, 3, 5)
    plt.plot(test_roc, label='Test ROC')
    plt.xlabel('Epoch')
    plt.ylabel('ROC')
    plt.title('Test ROC over Time')
    plt.legend()

    plt.tight_layout()
    plt.show()

# Example usage
test_results_path = r"C:\Users\Alex\Documents\BolT\Analysis\Logs\dynamic15Schaefer_bolT_hcpWM_20241202\test_results.pkl"
test_metrics_path = r"C:\Users\Alex\Documents\BolT\Analysis\Logs\dynamic15Schaefer_bolT_hcpWM_20241202\test_metrics.pkl"
plot_test_metrics(test_results_path, test_metrics_path)
```
