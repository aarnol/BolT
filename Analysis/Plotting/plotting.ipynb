```python
import os

# Define the path
path = r'C:\Users\Alex\Documents\BolT\loss_log.txt'

# Check if the file exists
if os.path.exists(path):
    print(f"The file exists at: {path}")
else:
    print(f"The file does not exist at: {path}")
```

```python
import numpy as np
import matplotlib.pyplot as plt 

# Load the data from the file
loss_values = np.loadtxt(path)

# Plot the loss values
plt.plot(loss_values)
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Loss over Epochs')
plt.show()
```

```python
# Trim the loss_values array to the largest multiple of 100
trimmed_size = (loss_values.size // 100) * 100
trimmed_loss_values = loss_values[:trimmed_size]

# Calculate the average of every 100 epochs
avg_loss_values = np.mean(trimmed_loss_values.reshape(-1, 100), axis=1)

# Plot the averaged loss values
plt.plot(avg_loss_values)
plt.xlabel('Epoch (in hundreds)')
plt.ylabel('Average Loss')
plt.title('Average Loss over Epochs')
plt.show()
```

```python
import pickle

# Define the path to the pickle file
pickle_path = r'C:\Users\Alex\Documents\BolT\step_metrics.pkl'

# Load the pickle file
with open(pickle_path, 'rb') as f:
    epoch_metrics = pickle.load(f)

# Print the contents of the pickle file
print(epoch_metrics)
```

```python
#plots for accuracy, precision, recall, roc, and loss over time
# Extract metrics from 
import matplotlib.pyplot as plt
import numpy as np

# Calculate the average of every 20 epochs

accuracy = [metric['accuracy'] for metric in epoch_metrics]
precision = [metric['precision'] for metric in epoch_metrics]
recall = [metric['recall'] for metric in epoch_metrics]
roc = [metric['roc'] for metric in epoch_metrics]
loss = [metric['loss'] for metric in epoch_metrics]


# Plot accuracy over time
plt.figure(figsize=(12, 6))
plt.subplot(2, 3, 1)
plt.plot(accuracy, label='Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Accuracy over Time')
plt.legend()

# Plot precision over time
plt.subplot(2, 3, 2)
plt.plot(precision, label='Precision')
plt.xlabel('Epoch')
plt.ylabel('Precision')
plt.title('Precision over Time')
plt.legend()

# Plot recall over time
plt.subplot(2, 3, 3)
plt.plot(recall, label='Recall')
plt.xlabel('Epoch')
plt.ylabel('Recall')
plt.title('Recall over Time')
plt.legend()

# Plot ROC over time
plt.subplot(2, 3, 4)
plt.plot(roc, label='ROC')
plt.xlabel('Epoch')
plt.ylabel('ROC')
plt.title('ROC over Time')
plt.legend()

# Plot loss over time
plt.subplot(2, 3, 5)
plt.plot(loss, label='Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Loss over Time')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# Trim the arrays to the largest multiple of 20
trimmed_size = (len(accuracy) // 20) * 20

avg_accuracy = np.mean(np.array(accuracy[:trimmed_size]).reshape(-1, 20), axis=1)
avg_precision = np.mean(np.array(precision[:trimmed_size]).reshape(-1, 20), axis=1)
avg_recall = np.mean(np.array(recall[:trimmed_size]).reshape(-1, 20), axis=1)
avg_roc = np.mean(np.array(roc[:trimmed_size]).reshape(-1, 20), axis=1)
avg_loss = np.mean(np.array(loss[:trimmed_size]).reshape(-1, 20), axis=1)
```

```python
# Plot averaged metrics over time
%matplotlib inline
plt.figure(figsize=(12, 6))

# Plot averaged accuracy over time
plt.subplot(2, 3, 1)
plt.plot(avg_accuracy, label='Avg Accuracy')
plt.xlabel('Step (in hundreds)')
plt.ylabel('Avg Accuracy')
plt.title('Average Accuracy over Time')
plt.legend()

# Plot averaged precision over time
plt.subplot(2, 3, 2)
plt.plot(avg_precision, label='Avg Precision')
plt.xlabel('Step (in hundreds)')
plt.ylabel('Avg Precision')
plt.title('Average Precision over Time')
plt.legend()

# Plot averaged recall over time
plt.subplot(2, 3, 3)
plt.plot(avg_recall, label='Avg Recall')
plt.xlabel('Step (in hundreds)')
plt.ylabel('Avg Recall')
plt.title('Average Recall over Time')
plt.legend()

# Plot averaged ROC over time
plt.subplot(2, 3, 4)
plt.plot(avg_roc, label='Avg ROC')
plt.xlabel('Step (in hundreds)')
plt.ylabel('Avg ROC')
plt.title('Average ROC over Time')
plt.legend()

# Plot averaged loss over time
plt.subplot(2, 3, 5)
plt.plot(avg_loss, label='Avg Loss')
plt.xlabel('Step (in hundreds)')
plt.ylabel('Avg Loss')
plt.title('Average Loss over Time')
plt.legend()

plt.tight_layout()
plt.show()
```
